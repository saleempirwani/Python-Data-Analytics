# Finance Data Analysis

```python
import datetime
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from pandas_datareader import wb, data

%matplotlib inline
```

```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
BAC = data.DataReader('BAC', 'stooq', start, end)

# Citi Group
C = data.DataReader('C', 'stooq', start, end)

# Goldman Sachs
GS = data.DataReader('GS', 'stooq', start, end)

# JPMorgan Chase
JPM = data.DataReader('JPM', 'stooq', start, end)

# Morgan Stanley
MS = data.DataReader('MS', 'stooq', start, end)

# Wells Fargo
WFC = data.DataReader('WFC', 'stooq', start, end)
```

### Create a list of ticker symbols (string) in alphabetical order

```python
tickets = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

### Use pd.concat(), to concat all banks dataframe together to all single dataframe.

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys=tickets)
```

```python
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
```

```python
bank_stocks.head()
```

## Exploratory data analysis

#### 1. What is the max Close price for each bank's stock throughout the time peroid? 

```python
bank_stocks.xs('Close', axis=1, level='Stock Info').max()
```

#### 2. Create a empty dataframe called returns.

```python
returns = pd.DataFrame()
```

```python
for tick in tickets:
    returns[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns
```

#### Create a pairplot of returns dataframe. What stock stands to you? Can you figure out why?

```python
sns.pairplot(returns)
```
