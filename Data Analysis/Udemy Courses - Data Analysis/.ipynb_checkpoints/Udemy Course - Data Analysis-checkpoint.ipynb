```python
import pandas as pd
```

### Data Pre-Processing

```python
# Convert 'price' column to int type.
df['price'] = df['price'].map(lambda x: '0' if x == 'Free' else x).astype('float64')
```

```python
# Convert 'timeStamp' column to datetime type.
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
```

## Exploratory data analysis

Q1. What are the different subjects/courses udemy offers?

```python
subjects = df['subject'].unique()

for sub in subjects:
    print(sub)
```

Q2. Find out the maximum no. of courses each category has?

```python
df['subject'].value_counts()
```

Q3. Show all the courses which are free of course?

```python
df[df['is_paid'] == False].head()
```

Q4. Show all the courses which are paid?

```python
df[df['is_paid'] == True].head()
```

Q5. Which are the top selling courses?

```python
df.sort_values('num_subscribers', ascending=False).head()
```

Q6. Which are the lease selling courses?

```python
df.sort_values('num_subscribers').head()
```

Q7. Show all the courses of graphic design, where the price is below 100?

```python
sub_mask = df['subject'] == 'Graphic Design'
price_mask1 = df['price'] > 0
price_mask2 = df['price'] < 100

df[sub_mask & price_mask1 & price_mask2]
```

Q8. Find out all the courses which are related to python?

```python
df[df['course_title'].str.contains('Python')].head()
```

Q9. What courses are published in 2017?

```python
df[ df['published_timestamp'].dt.year == 2017 ]
```

Q.10 What are the maximum no. of subscribers for each level of course?

```python
df.groupby('level')['num_subscribers'].sum().sort_values(ascending=False)
```
