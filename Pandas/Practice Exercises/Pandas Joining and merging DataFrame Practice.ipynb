# Pandas Joining and merging DataFrame Practice

````python
import pandas as pd
````

#### Dataset # 1

```    
student_data1:

student_id              name  marks
0         S1  Danniella Fenton    200
1         S2      Ryder Storey    210
2         S3      Bryce Jensen    190
3         S4         Ed Bernal    222
4         S5       Kwame Morin    199

student_data2:
  student_id              name  marks
0         S4  Scarlette Fisher    201
1         S5  Carla Williamson    200
2         S6       Dante Morse    198
3         S7    Kaiser William    219
4         S8   Madeeha Preston    201

```

````python
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})
````

#### Dataset # 2

```
data1:
  key1 key2   P   Q
0   K0   K0  P0  Q0
1   K0   K1  P1  Q1
2   K1   K0  P2  Q2
3   K2   K1  P3  Q3
data2:
  key1 key2   R   S
0   K0   K0  R0  S0
1   K1   K0  R1  S1
2   K1   K0  R2  S2
3   K2   K0  R3  S3
```

````python
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})
````

1. Write a Pandas program to join the two given dataframes along rows and assign all data. **Given Dataset1**.

````python
pd.concat([student_data1, student_data2])
````

2. Write a Pandas program to join the two given dataframes along columns and assign all data. **Given dataset # 1**

````python
pd.concat([student_data1, student_data2], axis=1)
````

3. Write a Pandas program to append rows to an existing DataFrame and display the combined data. **Dataset # 1 (student_data1)**

```
New Row(s)
student_id                  S6
name          Scarlette Fisher
marks                      205
dtype: object
```

````python
new_row = pd.Series(['S6', 'Scarlette Fisher', '205'], 
                    index=['student_id', 'name', 'marks'])

student_data1.append(new_row, ignore_index=True)
````

4. Write a Pandas program to append a list of dictioneries or series to a existing DataFrame and display the combined data. **Given Dataset # (student_data1)** 

````python
dicts = [{'student_id': 'S6', 'name': 'Scarlette Fisher', 'marks': 203},
         {'student_id': 'S7', 'name': 'Bryce Jensen', 'marks': 207}]

student_data1.append(dicts, ignore_index=True, sort=False)
````

5. Write a Pandas program to join the two given dataframes along rows and merge with another dataframe along the common column id.

````python
exam_data = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
        'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]})

result = pd.concat([student_data1, student_data2])

answer = pd.merge(result, exam_data, on="student_id")
answer
````

6. Write a Pandas program to join the two dataframes using the common column of both dataframes. 

````python
pd.merge(student_data1, student_data2, on="student_id", how='inner')
````

7. Write a Pandas program to join the two dataframes with matching records from both sides where available.

````python
pd.merge(student_data1, student_data2, on="student_id", how='outer')
````

8. Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only. **Given Dataset # 2**.

````python
pd.merge(data1, data2, on=['key1', 'key2'], how='left')
````

9. Write a Pandas program to join two dataframes using keys from right dataframe only.

````python
pd.merge(data1, data2, on=['key1', 'key2'], how='right')
````

````python
pd.merge(data2, data1, on=['key1', 'key2'], how='right')
````

10. Write a Pandas program to merge two given datasets using multiple join keys.

````python
pd.merge(data1, data2, on=['key1', 'key2'], how='inner')
````
