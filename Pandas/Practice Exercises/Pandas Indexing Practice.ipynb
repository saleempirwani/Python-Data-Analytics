10. Write a Pandas program to convert 1st and 3rd levels in the index into columns from a multiple level of index frame of a given dataframe.

```python
df2 = df.copy()

# index to multi-index
df2 = df2.set_index(['t_id', 'school_code', 'class'])

# resetting 2 index into columns
df2 = df2.reset_index(level=['t_id', 'class'])
df2
```
